```python
# Đọc file
df1=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_January_2019.csv')
df2=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_February_2019.csv')
df3=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_March_2019.csv')
df4=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_April_2019.csv')
df5=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_May_2019.csv')
df6=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_June_2019.csv')
df7=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_July_2019.csv')
df8=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_August_2019.csv')
df9=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_September_2019.csv')
df10=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_October_2019.csv')
df11=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_November_2019.csv')
df12=pd.read_csv('/content/drive/MyDrive/SALES PRODUCT DATA/Sales_December_2019.csv')
```

```python
# Gộp file
Sales_Data=pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12])
Sales=Sales_Data
```

```python
# Đọc DataFrame
Sales.info()
Sales.head()
```

```python
# Drop NA column
Sales = Sales.dropna(how = 'all').reset_index(drop=True)
```

```python
# Checking unique value của mỗi cột, Xem mỗi cột có bao nhiêu giá trị duy nhất
categorical=Sales.columns
for i in range(0,len(categorical)):
  print(categorical[i],": ",Sales[categorical[i]].nunique())
  print(Sales[categorical[i]].unique())
```

```python
# Bỏ giá trị Price Each và Quantity order không hợp lệ
Sales=Sales[Sales['Quantity Ordered']!='Quantity Ordered']
Sales=Sales[Sales['Price Each']!='Price Each']
Sales.info()
```

```python
# Gán giá trị số cho Quantity order and price each
Sales['Quantity Ordered']=Sales['Quantity Ordered'].astype(int)
Sales['Price Each']=Sales['Price Each'].astype(float)
# Chỉnh Order về dạng Datetime'
Sales['Order Date'] = pd.to_datetime(Sales['Order Date'])
```

```python
Sales['Order Date'].head()
```

```python
# Tạo côt revenue
Sales['Revenue']=Sales['Quantity Ordered']*Sales['Price Each']
# Tạo cột city
Sales["City"]=Sales['Purchase Address'].apply(lambda x: x.split(',')[1])
# Tạo cột Năm, Ngày, Tháng, Giờ, Phút
Sales['Date']=Sales['Order Date'].dt.day
Sales['Month']=Sales['Order Date'].dt.month
Sales['Year']=Sales['Order Date'].dt.year
Sales['Hour']=Sales['Order Date'].dt.hour
Sales['Minute']=Sales['Order Date'].dt.minute
# Bỏ năm 2020
Sales=Sales[Sales['Year']!=2020]
```

```python
# Sản phẩm bán cùng nhau
from itertools import combinations
from collections import Counter

together={}
# drop it using duplicated() funct
data = Sales[Sales['Order ID'].duplicated(keep=False)]
# create a new column
data['Grouped'] = Sales.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
# let's make a new variable
data = data[['Order ID', 'Grouped']].drop_duplicates()
# create a new variable for Counter
count = Counter()
# make a for loop
for row in data['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
```

```python
# create a DataFrame from the count dictionary
df = pd.DataFrame.from_dict(count, orient='index', columns=['Count'])

# reset the index and split the tuples into two columns
df = df.reset_index().rename(columns={'index': 'Product Pair'})
df[['Product 1', 'Product 2']] = pd.DataFrame(df['Product Pair'].tolist())

# sort the dataframe by the count column
df = df.sort_values(by='Count', ascending=False)

# show the first 10 rows
df=df.head(10)
Sales_together=df[['Product 1','Product 2','Count']]

Sales_together=Sales_together.reset_index(drop=True)
Sales_together
```
